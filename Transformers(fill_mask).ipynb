```python
import transformers
from transformers import pipeline
```

```python
mascarar = pipeline("fill-mask", model="neuralmind/bert-base-portuguese-cased")
```

```python
texto = mascarar (["[MASK] é a capital do Brasil", "A comida mais famosa do japão é a [MASK]"])
for x in range(len(texto)):
  print(texto[x])
```
